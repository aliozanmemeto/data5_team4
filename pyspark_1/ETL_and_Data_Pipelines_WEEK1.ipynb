```
# File location and type
file_location = "/FileStore/tables/2017_fordgobike_tripdata-2.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
df2 = df.filter(df.start_time > df.end_time)
```

```
from haversine import haversine
# Calculate the distance (in various units) between two points on Earth using their latitude and longitude.
#https://pypi.org/project/haversine/

#example
haversine((1,2)
          ,(3,4) )
```

```
from pyspark.sql.functions import udf

def haversine_f(a, b, c, d):
    return haversine((a,b),(c,d))
haversine_udf = udf(haversine_f)

df3 = df2.withColumn("distance", haversine_udf("start_station_latitude", "start_station_longitude", 
                                                 "end_station_latitude", "end_station_longitude"))
```
