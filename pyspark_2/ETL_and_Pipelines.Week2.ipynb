**Open the DataFrame**

```
# File location and type
file_location = "/FileStore/tables/2017_fordgobike_tripdataa.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

**Task 1 : Calculate the duration in seconds of each trip**

```
df.columns
```

```
## Duration functioin

def duration(h1,m1,s1,h2,m2,s2):
    sec1 = h1*3600 + m2*60 + s1
    sec2 = h2*3600 + m2*60 + s2
    
    return sec2 - sec1
```

```
##Apply the function

df = df.withColumn('duration_in_sec', duration(df[1], df[2], df[3], df[5], df[6], df[7]))
display( df[['start time hour', 'start time minute','start time seconds','end_time hour','end_time minute','end_time seconds','duration_in_sec']])
```

```
## Filter negative durations

df1 = df.filter(df['duration_in_sec'] > 0)
display( df1[['start time hour', 'start time minute','start time seconds','end_time hour','end_time minute','end_time seconds','duration_in_sec']])
```

**Task 2 : By assuming each minute cost 0.35 cent calculate the fee for each trip**

```
df1 = df1.withColumn("fee", (df1['duration_in_sec']/60)*0.35)
display(df1[['start_station_name','end_station_name','duration_in_sec','fee']])
```

**Task3 : Calculate the total distance for each bike and list the top 10**

```
display(df1)
```

```
from haversine import haversine 
from pyspark.sql.functions import udf
from pyspark.sql.types import DoubleType

def distance_calc(a,b,c,d):
    return haversine( (a,b), (c,d))

distance_udf = udf(distance_calc, DoubleType())

df1 = df1.withColumn('distance', distance_udf ( df1["start_station_latitude"], df1['start_station_longitude'],df1['end_station_latitude'], df1['end_station_longitude'] ) )

display(df1[["start_station_latitude", "start_station_longitude", "end_station_latitude", "end_station_longitude", "distance"]])
```

```
df_bike = df1.groupBy("bike_id").agg( {'distance' : 'sum'})
display(df_bike)
```

```
## Top 10 distances
from pyspark.sql.functions import desc
df_bike.orderBy( desc("sum(distance)")).show(10)
```
